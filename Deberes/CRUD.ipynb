```python
import json
class Auto:
    marca = None
    tipo = None
    anio = None
    precio = None
    transmision= None
    def __init__ (self, marca,tipo, anio, precio,transmision):
        self.marca = marca
        self.tipo = tipo
        self.anio = anio
        self.precio = precio
        self.transmision = transmision
    def __str__(self):
        return f"La marca del auto es: {self.marca} Tipo: {self.tipo} Año: {self.anio} Precio:{self.precio} Transmision: {self.transmision}"

    
def agregar_auto():
        marca = input("Ingrese la marca del auto con su serie ").upper()
        tipo = input ("Ingrese el tipo de auto ( Ej: Camioneta, Deportivo, Sedan) ").upper()
        anio = input("Ingrese el año del auto ")
        precio = input("Ingrese el precio del auto ")
        transmision = input("Ingrese que transmision posee el auto Ej: Automatica o Manual ").upper()
        autos.append(Auto(marca,tipo, anio, precio,transmision))
        print("Vehiculo ingresado con exito")
        
def actualizar_auto():
    marca=input("Ingresen la marca del auto que va a modificar ").upper()
    existe = False
    for i in autos:
        if(marca == i.marca):
            autos.remove(i)
            agregar_auto()
            existe = True
            break
    if(not existe):
        print("No existe el elemento")

def eliminar_auto():
    marca=input("Ingresen la marca del auto que va a \"Eliminar\" ").upper()
    existe = False
    for i in autos:
        if(marca== i.marca):
            autos.remove(i)
            existe = True
            break
    if(not existe):
        print("No existe la Marca")

        
def guardarAuto():
    try: 
        path = "./autosNuevo.txt"
        archivo_abierto = open(path, 'w') # w -> write
        guardar=""
        for auto_new in autos:
            guardar +=auto_new.marca+","+auto_new.tipo+","+auto_new.anio+","+auto_new.precio+","+auto_new.transmision
        archivo_abierto.write(guardar)
        archivo_abierto.close()
    except Exception as Error:
        print("Error guardar auto")
def imprimir_autos():
    for listar_autos in autos:
        print(listar_autos)
        

        
```

```python
class Partes:
    hp = None
    cilindraje = None
    version = None 
    tipo_motor = None
    refrigeracion = None 
    numero_de_marchas = None
    def __init__(self,nombre,hp, cilindraje,version,tipo_motor,refrigeracion,numero_de_marchas):
        self.nombre = nombre
        self.hp = hp
        self.cilindraje = cilindraje
        self.version = version
        self.tipo_motor = tipo_motor
        self.refrigeracion = refrigeracion
        self.numero_de_marchas = numero_de_marchas
    def __str__(self):
        return f"Nombre del Auto:{self.nombre} Horse Power: {self.hp} Cilindraje: {self.cilindraje} Version: {self.version} Tipo Motor: {self.tipo_motor} Refrigeracion: {self.refrigeracion} Numero Marchas: {self.numero_de_marchas}"

    
def agregar_partes():
        nombre = str(input("Ingrese el nombre del auto al que desea especificar las partes")).upper()
        for buscar in autos:
            if (buscar.marca == nombre):
                hp = input("Ingrese el horse power del vehiculo ")
                cilindraje = input ("Ingrese el cilindraje ")
                version = input("Ingrese la version del auto ").upper()
                tipo_motor = input("Ingrese el tipo de motor ")
                refrigeracion = input("Ingrese la refrigeracion Ej: Naturalmente Aspirada , Turbo").upper()
                numero_de_marchas = input("Ingrese el numero de marchas del auto ")
                partes.append(Partes(nombre,hp, cilindraje,version,tipo_motor,refrigeracion,numero_de_marchas))
            else:
                print("No existe un registro del auto ingresado")
        
        
def actualizar_partes():
    nombre=input("Ingresen el nombre del auto que va a modificar las partes ").upper()
    existe=False
    for i in partes:
        if(nombre == i.nombre):
            partes.remove(i)
            agregar_partes()
            existe=True
            break
    if(not existe):
        print("No existe el auto que se desea modificar")
def eliminar_partes():
    nombre=input("Ingresen la marca del auto que va a \"Eliminar\" ").upper()
    existe=False
    for i in partes:
        if(nombre== i.nombre):
            partes.remove(i)
            existe=True
            break
    if(not existe):
        print("No existe el auto")

        
def guardarPartes():
    try: 
        path = "./partes.txt"
        archivo_abierto = open(path, 'w') # w -> write
        guardar=""
        for partes_new in partes:
            guardar +=partes_new.nombre+","+partes_new.hp+","+partes_new.cilindraje+","+partes_new.version+","+partes_new.tipo_motor+","+partes_new.numero_de_marchas
        archivo_abierto.write(guardar)
        archivo_abierto.close()
    except Exception as Error:
        print("Error al guardar las partes")

def imprimir_partes():
    for listar_partes in partes:
        print(listar_partes)
```

```python
autos = []
partes = []
while True:
    print("****Bienvenido al consecionario Autoland****")
    print("Ingrese el numero de la opcion que desea realizar")
    opciones = str(input ("1 Ingresar un nuevo auto \n2 Listar autos \n3 Actualizar un Auto \n4 Eliminar un auto \n5 Ingresar las partes de un auto \n6 Listar partes \n7 Actualizar Partes \n8 Eliminar Partes \n9 Salir"))
    if opciones == "1":
        print ("Ha seleccionado agergar un Auto") 
        agregar_auto()
        guardarAuto() 
    elif opciones == "2" :
        imprimir_autos()
    elif opciones == "3":
        imprimir_autos()
        actualizar_auto()
        guardarAuto()
    elif opciones=="4":
        imprimir_autos()
        eliminar_auto()
    elif opciones=="5":
        agregar_partes()
        guardarPartes()
    elif opciones == "6":
        imprimir_partes()
    elif opciones == "7":
        imprimir_partes()
        actualizar_partes()
    elif opciones == "8":
        eliminar_partes()
    elif opciones=="9":
        print("*******Adios*******")
        break
```
