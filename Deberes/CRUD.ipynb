```python
import json
class Auto:
    marca = None
    tipo = None
    anio = None
    precio = None
    transmision= None
    def __init__ (self, marca,tipo, anio, precio,transmision):
        self.marca = marca
        self.tipo = tipo
        self.anio = anio
        self.precio = precio
        self.transmision = transmision
    def __str__(self):
        return f"Marca: {self.marca} Tipo: {self.tipo} Año: {self.anio} Precio:{self.precio} Transmision: {self.transmision}"

    
def agregar_auto():
        #autos = []
        marca = input("Ingrese la marca del auto").upper()
        tipo = input ("Ingrese el tipo de auto Ej: Camioneta, Deportivo, Sedan").upper()
        anio = input("Ingrese el año del auto ")
        precio = input("Ingrese el precio del auto ")
        transmision = input("Ingrese que transmision posee el auto Ej: Automatica o Manual ").upper()
        autos.append(Auto(marca,tipo, anio, precio,transmision))
        
def actualizar_auto():
    marca=input("Ingresen la marca del auto que va a modificar ").upper()
    estado=False
    for i in autos:
        if(marca == i.marca):
            autos.remove(i)
            agregar_auto()
            estado=True
            break
    if(not estado):
        print("No existe el elemento")
def eliminar_auto():
    marca=input("Ingresen la marca del auto que va a \"Eliminar\" ").upper()
    estado=False
    for i in autos:
        if(marca== i.marca):
            autos.remove(i)
            estado=True
            break
    if(not estado):
        print("No existe la Marca")

        
def guardarAuto():
    try: 
        path = "./autosNuevo.txt"
        archivo_abierto = open(path, 'w') # w -> write
        guardar=""
        for auto_new in autos:
            guardar +=auto_new.marca+"||"+auto_new.tipo+"||"+auto_new.anio+"||"+auto_new.precio+"||"+auto_new.transmision+"||"+"\n"
        archivo_abierto.write(guardar)
        archivo_abierto.close()
    except Exception as Error:
        print("Error guardar auto")
def imprimir_autos():
    for listar_autos in autos:
        print(listar_autos)
        
```

```python
autos=[]
while True:
    print("Bienvenido al consecionario Autoland")
    print("Ingrese el numero de la opcion que desea realizar")
    opciones = str(input ("1 Ingresar un nuevo auto \n2 Listar autos \n3 Actualizar un Auto \n4 Eliminar un auto \n5 Salir"))
    if opciones == "1":
        print ("Ha seleccionado agergar un Auto") 
        agregar_auto()
        guardarAuto() 
    elif opciones == "2" :
        imprimir_autos()
    elif opciones == "3":
        imprimir_autos()
        actualizar_auto()
        guardarAuto()
    elif opciones=="4":
        imprimir_autos()
        eliminar_auto()
    elif opciones=="5":
        print("Adios")
        break
```
